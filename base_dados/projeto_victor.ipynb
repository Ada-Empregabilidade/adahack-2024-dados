```python
df=df.drop('id.1', axis=1) #drop da coluna id duplicada
```

```python
df=df.drop('EndereÃ§o', axis=1) #drop da coluna endereço sem uso
```

```python
df.rename(columns={
    'Raça': 'raca',
    'Formação': 'formacao',
    'Tempo de Casa': 'tempo_de_casa',
    'senioridade': 'senioridade'
}, inplace=True)
#renomear as colunas
```

```python
display(df)
```

```python
mapeamento = {
    'Ã©': 'é',
    'Ã¢': 'â',
    'Ãº': 'ú',
    'Ã­': 'í',
    'Ã§': 'ç',
    'Ã¡': 'á',
    'Ã³': 'ó',
    'Ãª': 'ê',
    'Ã®': 'î',
    'Ã´': 'ô',
    'Ã»': 'û',
    'Ã£': 'ã',
    'Ãµ': 'õ',
    'Ã ': 'à',
    'Ã¨': 'è',
    'Ã¬': 'ì',
    'Ã¹': 'ù',
    'Ã¤': 'ä',
    'Ã«': 'ë',
    'Ã¯': 'ï',
    'Ã¶': 'ö',
    'Ã¼': 'ü'
}

for col in df.columns:
    df[col] = df[col].replace(mapeamento, regex=True)  #ajuste dos dados
```

```python
display(df)
```

```python
df.describe().T  #visualização estatística da base
```

```python
df['tempo_de_casa'] = df['tempo_de_casa'].abs() #retirada de valores nulo. No caso tempo de casão não faz sentido ser negativo.
```

```python
df.describe().T #Confirmação
```

```python
df_menor_18 = df[df['Idade'] < 18] #Busca de dados incoerentes: Pessoas menores de 18 anos x Vaga e Cargos 
```

```python
display(df_menor_18)
```

```python
df['idadextempodecasa'] = df['Idade'] - df['tempo_de_casa']   #Verificando dados coesos entre idade e tempo de casa
```

```python
display(df)
```

```python
df.describe().T
```

```python
df_tempo_de_casa = df[df['idadextempodecasa'] < 0]
```

```python
display(df_tempo_de_casa) #encontrei uma pessoa menor de idade, mesmo na legislação vigente, com tempo de casa maior que de vida
```
