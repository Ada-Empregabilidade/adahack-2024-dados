```python
#Excluir coluna id duplicada

base_dados = base_dados.drop('id.1', axis=1)

base_dados
```

```python
#Corrigir os nomes das colunas do Dataframe

# Criar um dicionário com os nomes antigos das colunas como chaves e os nomes novos como valores
novo_nomes_colunas = {col: substituir_codificados(col) for col in base_dados.columns}

# Renomear as colunas do DataFrame usando o dicionário
base_dados = base_dados.rename(columns=novo_nomes_colunas)

# Imprimir os novos nomes das colunas
print(base_dados.columns)
```

```python
base_dados.dtypes
```

```python
#Formação apresenta dados em float, vamos corrigir isso

base_dados['Formação'] = base_dados['Formação'].astype(str)

```

```python
#Separar as colunas que contem somente texto
colunas_texto = ['Nome', 'Genero', 'Raça', 'Endereço', 'Estado', 'Formação', 'Departamento', 'Senioridade']

# Aplicar a função substituir_codificados a cada elemento do DataFrame
for item in colunas_texto:
   base_dados[item] = base_dados[item].apply(substituir_codificados)

# Imprimir o DataFrame para verificar os resultados
print(base_dados)
```
