```python
# Função para resumo
def resumo_tabela(df):
    print(f"Dataset Shape: {df.shape}")
    sumario = pd.DataFrame(df.dtypes,columns=['dtypes'])
    sumario = sumario.reset_index()
    sumario['Name'] = sumario['index']
    sumario = sumario[['Name','dtypes']]
    sumario['Missing'] = df.isnull().sum().values    
    sumario['Uniques'] = df.nunique().values
    sumario['First Value'] = df.loc[0].values
    sumario['Last Value'] = df.iloc[-1, :].values
    

    for name in sumario['Name'].value_counts().index:
        sumario.loc[sumario['Name'] == name, 'Entropy'] = round(stats.entropy(df[name].value_counts(normalize=True), base=2),2) 

    return sumario
```

# Código para substituir idade

```python
# Calcular a mediana por senioridade
mediana_por_senioridade = df_empresa.groupby('Senioridade')['Idade'].median()

# Substituir os valores nulos e menores que 16 pela mediana de cada senioridade
for senioridade, mediana in mediana_por_senioridade.items():
    mascara = (df_empresa['Senioridade'] == senioridade) & ((df_empresa['Idade'].isnull()) | (df_empresa['Idade'] < 16))
    df_empresa.loc[mascara, 'Idade'] = mediana

# Exibir o DataFrame após as substituições
print(df_empresa)

```
