### Vamos corrigir os erros de escrita:

Vendo nosso dataframe acima já conseguimos ver visualmente vários erros de escrita que podemos corrigir, infelizmente o enconding que era para ajudar a gente nesta parte não funcionou então foi implementado manualmente e ao longo da nossa ánalise caso algo tenha faltado nós iremos corrigir.

```python
# Função para substituir padrões de escrita em títulos das colunas e valores das células
def substituir_padroes(texto):
    if isinstance(texto, str):
        correcoes = {
            'Ã§': 'ç',
            'Ã¡': 'á',
            'Ã£': 'ã',
            'Ã©': 'é',
            'Ã³': 'ó',
            'Ã´': 'ô',
            'Ãº': 'ú',
            'Ãª': 'ê',
            'Ã‚': 'Â',
            'Ã¢': 'â',
            'Ã': 'í',
            'íµ': 'õ',
            'í': 'Á',
            'í´': 'ô'
        }
        for errado, correto in correcoes.items():
            texto = texto.replace(errado, correto)
    return texto

# Aplicar a função nos títulos das colunas
df.columns = df.columns.map(substituir_padroes)

# Aplicar a função em todo o DataFrame (após corrigir os títulos das colunas)
df = df.applymap(substituir_padroes)

# Exibir o DataFrame resultante
df.head(10)
```

##### A quantidade de valores negativos na nossa base é aproximadamente 1% dos valores, nos podemos intepretar de algumas formas a correção destes dados.
1. Podemos interpretar como um erro de digitação no '-', neste caso nos iriamos substituir pelos mesmos valores mas positivos.
2. Se a distribuição dos dados for aproximadamente simétrica (ou seja, não houver viés significativo para a esquerda ou para a direita), substituir os valores negativos pela média pode ser a melhor escolha.
3. Se os dados forem enviesados (por exemplo, se houver uma cauda longa para a direita ou para a esquerda), substituir os valores negativos pela mediana pode ser mais apropriado, pois a mediana é menos sensível a valores extremos do que a média.

```python
# Calculando a contagem dos valores na coluna 'Tempo de casa'
contagem_tempo_casa = df['Tempo de casa'].value_counts()

# Ordenando os valores por índice
contagem_tempo_casa = contagem_tempo_casa.sort_index()

# Plotando o gráfico de barras
contagem_tempo_casa.plot.bar(color='skyblue', edgecolor='black')

# Adicionando rótulos e título ao gráfico
plt.xlabel('Tempo de casa')
plt.ylabel('Contagem')
plt.title('Contagem de Tempo de Casa')

# Calculando a assimetria (skewness) e a curtose (kurtosis)
skewness = df['Tempo de casa'].skew()
kurtosis = df['Tempo de casa'].kurtosis()

# Exibindo os valores de skewness e kurtosis
plt.text(0.05, 0.5, f'Skewness: {skewness:.2f}\nKurtosis: {kurtosis:.2f}', transform=plt.gca().transAxes)

# Exibindo o gráfico
plt.show()
```

Com base nesses valores, como a distribuição é aproximadamente simétrica (skewness próxima de 0) e um pouco achatada (kurtosis negativa), tanto a média quanto a mediana podem ser opções razoáveis para substituir os valores ausentes. No entanto, se houver a presença de valores discrepantes (outliers), a mediana pode ser mais robusta, pois é menos sensível a esses valores extremos. De acordo com isso decidimos substituir os valores nulos pela mediana.

```python
# Calculando a mediana
mediana = df['Tempo de casa'].median()

# Substituindo os valores negativos pela mediana
df.loc[df['Tempo de casa'] < 0, 'Tempo de casa'] = mediana
```

### **Criação da Coluna 'Faixa de Idade'**

Vamos criar uma coluna 'Faixa de Idade' para categorizar melhor os funcionários e gerar novos insights. Essa coluna irá agrupar as idades em faixas específicas, como 0 a 20 anos, 20 a 30 anos, 31 a 40 anos, 41 a 50 anos e acima de 50 anos.
