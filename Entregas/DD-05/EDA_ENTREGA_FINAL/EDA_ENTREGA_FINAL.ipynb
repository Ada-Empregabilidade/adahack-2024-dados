## **Metodologia**

>Para executar o trabalho, o grupo se baseou na metodologia CRISP-DM. A sigla significa *Cross Industry Standard Process for Data Mining*, traduzido como "Processo Padrão Inter-Indústrias para Mineração de Dados".
A metodologia citada anteriormente é constituida pelas seguintes etapas:

* **Business Understanding**: A primeira etapa trata de entender o contexto da empresa e quais os desafios a serem resolvidos.

* **Data Understanding**: É a fase de se fazer o estudo e levantamento dos conteúdos, conseguindo mapear se existem anomalias, quais os dados faltantes e como será a extração.

* **Data Preparation**: É onde vai organizar quais os dados que serão modelados para o projeto, garantir que as informações estão alinhadas de acordo com o formato alinhado e o que será relevante para cruzar nas informações.

* **Modeling**: Essa é a fase de modelar. Depois que foi feita a mineração inicial, limpeza e tratamento dos dados, o próximo passo é desenvolver o  modelo.

* **Evaluation**: Após a criação de modelos e algoritmos, nessa etapa será feita a avaliação dos resultados e quais as variações podem surgir. É importante observar se o modelo criado atende aos objetivos iniciais traçados na etapa de entendimento do negócio.

* **Deployment**: Nesta etapa, será implementado o protótipo para que o cliente faça o acompanhamento de sua performance após sua entrega.

#### **2.4) Correção de formato dos dados textuais**
>Mesmo depois de testados alguns encodings ('utf-8', 'latin-1', 'iso-8859-15'), ficou perceptível que todos os utilizados ainda apresentam erros de digitação.Portanto, será implementada uma correção para esses dados.

```python
df.info()
```

Vendo o dataframe acima já fica notório  vários erros de escrita que podem ser corrigidos, infelizmente o enconding  não funcionou então foi implementado manualmente e ao longo da  ánalise caso algo tenha faltado será corrigido.

```python
# Função para substituir padrões de escrita em títulos das colunas e valores das células
def substituir_padroes(texto):
  """
  Função para substituir padrões de escrita incorreta em títulos de colunas
  e valores das células.
  """
  if isinstance(texto, str):
        correcoes = {
            'Ã§': 'ç',
            'Ã¡': 'á',
            'Ã£': 'ã',
            'Ã©': 'é',
            'Ã³': 'ó',
            'Ã´': 'ô',
            'Ãº': 'ú',
            'Ãª': 'ê',
            'Ã‚': 'Â',
            'Ã¢': 'â',
            'Ã': 'í',
            'íµ': 'õ',
            'í': 'Á',
            'í´': 'ô'
        }
        for errado, correto in correcoes.items():
            texto = texto.replace(errado, correto)
  return texto

# Aplicar a função nos títulos das colunas
df.columns = df.columns.map(substituir_padroes)

# Aplicar a função em todo o DataFrame (após corrigir os títulos das colunas)
df = df.applymap(substituir_padroes)

# Exibir o DataFrame resultante
df.head(10)
```

No entanto, se houver a presença de valores discrepantes (outliers), a mediana pode ser mais robusta, pois é menos sensível a esses valores extremos.
```python
# Verificando qual a mediana da idade de cada senioridade
mediana_por_senioridade = df.groupby('senioridade')['idade'].median()
```

```python
# substuindo idades nulas e idades menores que 18 pela mediana de idades por senioridade
for senioridade, mediana in mediana_por_senioridade.items():
    mascara = (df['senioridade'] == senioridade) & ((df['idade'].isnull()) | (df['idade'] < 18))
    df.loc[mascara, 'idade'] = mediana
```

2. Se a distribuição dos dados for aproximadamente simétrica (ou seja, não houver viés significativo para a esquerda ou para a direita), substituir os valores negativos pela média pode ser a melhor escolha.
3. Se os dados forem enviesados (por exemplo, se houver uma cauda longa para a direita ou para a esquerda), substituir os valores negativos pela mediana pode ser mais apropriado, pois a mediana é menos sensível a valores extremos do que a média.

##### **4.4.7) Calculando a dissimilaridade  entre as proporções de raça e genero para as diferentes bases**

```python
# calculando índice de dissimilaridade


# Dados do IBGE
# Raças e suas respectivas porcentagens na população brasileira
racas = ['Pardo', 'Preto', 'Branco', 'Amarelo', 'Indígena']
dados_IBGE = [45.30, 10.20, 43.50, 0.40, 0.60]

# Dados da empresa - Raça
# Porcentagens de cada raça para diferentes níveis de cargo na empresa
dados_empresa_raca = {
    'Pardo': [47.54, 45.88, 46.23, 49.94],  # Júnior, Pleno, Sênior, Gestor
    'Preto': [9.31, 8.64, 9.15, 7.93],       # Júnior, Pleno, Sênior, Gestor
    'Branco': [42.28, 44.55, 43.17, 41.31],  # Júnior, Pleno, Sênior, Gestor
    'Amarelo': [0.44, 0.4, 0.55, 0.57],      # Júnior, Pleno, Sênior, Gestor
    'Indígena': [0.44, 0.4, 0.55, 0.57]      # Júnior, Pleno, Sênior, Gestor
}

# Dados do IBGE - Gênero
# Porcentagens de gênero na população brasileira
porcentagem_feminino_IBGE = 51.50
porcentagem_masculino_IBGE = 48.50

# Dados da empresa - Gênero
# Porcentagens de gênero para diferentes níveis de cargo na empresa
dados_empresa_genero = {
    'Pleno': [25.73, 74.27],  # Feminino, Masculino
    'Júnior': [25.96, 74.04], # Feminino, Masculino
    'Gestor': [23.68, 76.32], # Feminino, Masculino
    'Sênior': [24.77, 75.23]  # Feminino, Masculino
}

# Calcular as proporções de raça da empresa e do IBGE
# Transpor a matriz para que as dimensões coincidam
proporcoes_empresa_raca = np.array([dados_empresa_raca[raca] for raca in racas]).T / 100
proporcoes_IBGE_raca = np.array(dados_IBGE) / 100

# Calcular as proporções de gênero da empresa e do IBGE
proporcoes_empresa_genero = np.array([dados_empresa_genero[cargo] for cargo in dados_empresa_genero]) / 100
proporcoes_IBGE_genero = np.array([porcentagem_feminino_IBGE, porcentagem_masculino_IBGE]) / 100

# Calcular a diferença média absoluta para raça
diff_raca = np.abs(proporcoes_empresa_raca - proporcoes_IBGE_raca).mean()

# Calcular a diferença média absoluta para gênero
diff_genero = np.abs(proporcoes_empresa_genero - proporcoes_IBGE_genero).mean()

# Calcular o Índice de Dissimilaridade
indice_dissimilaridade = (diff_raca + diff_genero) / 2

# Imprimir o Índice de Dissimilaridade
print("Índice de Dissimilaridade:", indice_dissimilaridade)
```
